```python
def f(x) :
    if 0 <= x <= 10:
        return -0.005 * x + 1
    elif 10 < x <= 105:
        return -0.01 * x + 1.05
    else:
        return 0
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = [_ for _ in range(150)]
y = [f(_) for _ in x]

plt.figure(1)
plt.plot(x, y)
```
